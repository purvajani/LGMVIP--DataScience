Task 2 : Image to Pencil Sketch with Python

**Import Libraries**

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

**Read Photo**

```python
from google.colab import files
uploaded= files.upload()
```

```python
image = cv2.imread("/content/virat-kohli.jpg")
image
     
```

**Showing Image**

```python
cv2_imshow(image)
```

**Converting Image into Gray Scale Image**

```python
gray_scale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_scale_image)
```

**Inverting the gray scale image**

```python
inverted_image = cv2.bitwise_not(gray_scale_image)
cv2_imshow(inverted_image)
```

**Image Smoothing**

```python
img_smoothing = cv2.GaussianBlur(inverted_image, (21, 21), sigmaX=0, sigmaY=0)
cv2_imshow(img_smoothing)
```

**Final Step**

```python
def pencil_sketch(x, y):
  return cv2.divide(x, 255-y, scale=256)     
```

```python
pencilsketch = pencil_sketch(gray_scale_image, img_smoothing)
cv2_imshow(pencilsketch)
```
